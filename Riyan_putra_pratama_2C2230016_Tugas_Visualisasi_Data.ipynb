**Import library**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Import Data set**

```python
data = pd.read_csv('/content/drive/MyDrive/visualisasi data python/H05.csv')
data
```

**1.Histogram distribusi nilai matematika**

```python
counts, bins, patches = plt.hist(data['nilai_mat'], bins=10, edgecolor='black', color='skyblue')
patches[counts.argmax()].set_facecolor('red')  # bar tertinggi dirubah warna nya jadi merah sesuai frekuensi tertinggi
plt.title('Distribusi Nilai Matematika')
plt.xlabel('Nilai Matematika')
plt.ylabel('Frekuensi')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

 **Keterangan:**

 Barchat diatas menunjukann chart tertinggi berada pada interval 60 â€“ 70, maka kebanyakan mahasiswa mendapatkan nilai matematika dalam rentang tersebut. Bar tertinggi diberi warna merah untuk menonjolkan rentang nilai yang paling banyak dihuni mahasiswa

**2.Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas.**

```python
fakultas = data['fakultas'].value_counts()
colors = ['red' if i == fakultas.argmax() else 'blue' for i in range(len(fakultas))]
fakultas.plot(kind='barh', color=colors, edgecolor='black', title='Jumlah Mahasiswa setiap Fakultas', figsize=(8, 6))
plt.xlabel('Jumlah Mahasiswa')
plt.ylabel('Fakultas')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

**Keterangan:**

Horizontal bar chat di atas menunjukkan jumlah mahasiswa dari setiap fakultas, dan menunjukan bahwa FTI mempunyai mahasiswa terbanyak yang ditunjukan dengan chart paling tinggi dan mempunyai warna merah

**3.Diagram pie banyaknya mahasiswa masing-masing fakultas**

```python
data['fakultas'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140, figsize=(8, 8), title='Proporsi Mahasiswa setiap Fakultas')
plt.ylabel('')
plt.show()
```

Menampilkan proporsi mahasiswa di tiap fakultas dalam bentuk persentase.

**5. Scatter plot dengan nilai kimia sebagai x dan fisika sebagai y.**

```python
data.plot(kind='scatter', x='nilai_kim', y='nilai_fis', alpha=0.7, color='cyan', edgecolor='black', title='Scatter Plot Nilai Kimia vs Nilai Fisika', figsize=(8, 6))
plt.xlabel('Nilai Kimia')
plt.ylabel('Nilai Fisika')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

**keterangan:**

Dalam scatter plot ini menunjukan bahwa ada hubungan positif antara nilai kimia dan nilai fisika yang ditandakan dengan titik titik pada sumbu x dan y yang membentuk pola naik ke atas
