Problem (1) Membuat program menerima bilangan N dan menuliskan bilangan 10x terkecil yang lebih dari N

```python

n = int(input("Masukkan N: "))

x = 0 #pangkat
while 10  **  x <= n:
  x += 1

print(10 ** x)
```

Problem (2) buat program yang menerima N dan membuat pola segitiga angka dengan dasar N

```python
n = int(input("Masukkan nilai n:"))

for i in range(n ,0 ,-1):
  for j in range(1, i+1):
    print(j, end="")
  print()
```

Problem (3) Kalkulator sederhana

```python
while True:
  angka_pertama = int(input("Masukkan angka pertama: "))
  angka_kedua = int(input("Masukkan angka kedua: "))
  operator = input("Masukkan operator (+, -, *, /, %): ")

  if operator == "+":
    hasil = angka_pertama + angka_kedua
  elif operator == "-":
    hasil = angka_pertama - angka_kedua
  elif operator == "*":
    hasil = angka_pertama * angka_kedua
  elif operator == "/":
    if angka_kedua == 0:
      print("Tidak dapat dibagi dengan nol.")
      continue
    else:
      hasil = angka_pertama // angka_kedua
  elif operator == "%":
    if angka_kedua == 0:
      print("Tidak dapat dibagi dengan nol.")
      continue
    else:
      hasil = angka_pertama % angka_kedua
  else:
    print("Operator tidak valid.")
    continue

  print(f"{angka_pertama} {operator} {angka_kedua} = {hasil}")

  lanjutkan = input("Apakah Anda ingin menghitung lagi? (y/n): ")
  if lanjutkan.lower() != 'y':
    break
```

Problrm(4) bilangan ganjil

```python
angka_awal = int(input("Masukkan angka awal: "))
angka_akhir = int(input("Masukkan angka akhir: "))
x = []
y = []

for i in range(angka_awal, angka_akhir+1):
  if i % 2 == 0:
    x.append(i)
  else:
    y.append(i)

print("Bilangan genap:", x)
print("Bilangan ganjil:", y)
```

Problem (1)

```python
N=int(input("Masukkan nilai N :"))
A=int(input("Masukkan nilai A :"))
B=int(input("Masukkan nilai B :"))
C=int(input("Masukkan nilai C :"))

for i in range(1, N+1):
  if i % A == 0 and i % B == 0 and i % C == 0:
    print("SiapBangJago", end=" ")
  elif i % A == 0 and i % B == 0:
    print("SiapBang", end=" ")
  elif i % A == 0 and i % C == 0:
    print("SiapJago", end=" ")
  elif i % B == 0 and i % C == 0:
    print("BangJago", end=" ")
  elif i % A == 0:
    print("Siap", end=" ")
  elif i % B == 0:
    print("Bang", end=" ")
  elif i % C == 0:
    print("Jago", end=" ")
  else:
    print(i, end=" ")

print()
```

- Interpretasi coding:
1. Bagian input :


*   Kode meminta pengguna untuk memasukkan empat nilai integer: N, A, B, dan C.
* N menentukan batas atas dari angka yang akan diproses, yaitu dari 1 hingga N.  
* A, B, dan C adalah bilangan yang digunakan untuk memeriksa kelipatan.

2. Pengulangan

* Menggunakan loop for, kode ini akan memeriksa setiap angka i dari 1 hingga N.
* Pada setiap iterasi, kode memeriksa beberapa kondisi menggunakan pernyataan if.

3. pengkondisian
* Jika i habis dibagi ketiga bilangan tersebut, maka akan mencetak "SiapBangJago".
* Jika i habis dibagi A dan B, maka akan mencetak "SiapBang".
* Jika i habis dibagi A dan C, maka akan mencetak "SiapJago".
* Jika i habis dibagi B dan C, maka akan mencetak "BangJago".
* Jika i hanya habis dibagi A, maka akan mencetak "Siap".
* Jika i hanya habis dibagi B, maka akan mencetak "Bang".
* Jika i hanya habis dibagi C, maka akan mencetak "Jago".
* Jika tidak ada kondisi yang terpenuhi, maka angka i akan dicetak.
