FOR NESTED LOOP

```python
n = int(input("Masukkan nilai n:"))

for i in range(n):
  for j in range(n):
    print("*", end="|") # diekesekusi sebanyak n * n kali

    #dieksekusi sebanyak n kali
    #perhatikan indentasi di dalam perulangan.
```

```python
n = int(input("Masukkan nilai n:"))

for i in range(n ,0 ,-1):
  for j in range(1, i+1):
    print(j, end="")
  print()
```

```python
n = int(input("Masukkan nilai n:"))

kelipatan7 =(n // 7 + 1) * 7
for i in range(7, kelipatan7, 7):
  print(i)
```

```python
n = int(input("Masukkan nilai n:"))

for i in range(1, n+1):
  if i % 2 == 0 and i % 3 == 0:
    print("Dua tiga", end=" ")
  elif i % 2 == 0:
    print("Dua", end="")
  elif i % 3 == 0:
    print("tiga", end="")
  else:
    print(i, end=" ")
print()
```
