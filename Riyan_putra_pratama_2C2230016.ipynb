**NO 2**

Buatlah algoritma/program yang membaca sebuah nilai integer positif, misalnya N, dan menjumlahkan (serta menampilkan) semua bilangan kelipatan 5 antara 1 s.d. N.

```python
# Program untuk menghitung jumlah bilangan kelipatan 5 antara 1 sampai N

# Input nilai N
N = int(input("Masukkan nilai N: "))

# Inisialisasi variabel untuk menyimpan total jumlah
total = 0

# Menyimpan bilangan kelipatan 5 dalam sebuah list untuk ditampilkan nanti
kelipatan_5 = []

# Perulangan untuk mencari kelipatan 5 dari 1 sampai N
for i in range(1, N+1):
    if i % 5 == 0:
        kelipatan_5.append(i)  # Tambahkan bilangan ke list kelipatan_5
        total += i             # Tambahkan bilangan ke total

# Output hasil
if kelipatan_5:
    print("Bilangan kelipatan 5 antara 1 sampai", N, "adalah:", kelipatan_5)
    print("Jumlah total kelipatan 5 tersebut adalah:", total)
else:
    print("Tidak ada bilangan kelipatan 5 antara 1 sampai", N)
```

**NO 3**

Buatlah 3 versi program yang menerima masukan sebuah integer positif, misalnya N (asumsi N>0), dan menuliskan lirik lagu Anak Ayam di atas dengan menggunakan perulangan for dan while.

A. perulangan for


```python
# Versi menggunakan for loop
N = int(input("Masukkan jumlah anak ayam: "))

print(f"Anak ayam turunlah {N}")
for i in range(N, 1, -1):
    print(f"Mati satu tinggallah {i - 1}")
print("Mati satu tinggallah induknya")
```

B.Perulangan while

```python
# Versi menggunakan while loop
N = int(input("Masukkan jumlah anak ayam: "))

print(f"Anak ayam turunlah {N}")
while N > 1:
    N -= 1
    print(f"Mati satu tinggallah {N}")
print("Mati satu tinggallah induknya")
```
